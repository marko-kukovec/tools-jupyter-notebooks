Endpoint docs
https://www.nicehash.com/software-developers

````python
import requests
import pandas as pd

multialgo_response = requests.get('https://api2.nicehash.com/main/api/v2/public/simplemultialgo/info')
multialgo_response.raise_for_status()
multialgo_data = multialgo_response.json()
algo_data = { element['algorithm']: {'paying': element['paying'], 'speed': element['speed']} for element in multialgo_data['miningAlgorithms'] }

algorithms_response = requests.get('https://api2.nicehash.com/main/api/v2/mining/algorithms')
algorithms_response.raise_for_status()
algorithms_data = algorithms_response.json()

index = [ element['algorithm'] for element in algorithms_data['miningAlgorithms'] ]
columns = ['marketFactor', 'displayMarketFactor', 'priceFactor', 'displayPriceFactor', 'miningFactor', 'displayMiningFactor', 'priceDownStep', 'paying', 'speed']
data = [ [ float(element['marketFactor']), element['displayMarketFactor'], float(element['priceFactor']), element['displayPriceFactor'], float(element['miningFactor']), element['displayMiningFactor'], float(element['priceDownStep']), algo_data[element['algorithm']]['paying'], algo_data[element['algorithm']]['speed'] ] for element in algorithms_data['miningAlgorithms'] ]

raw_data_df = pd.DataFrame(data, index=index, columns=columns)
raw_data_df
````

````python
calc_df = pd.DataFrame(index=index)
sats2btc = 1e8

calc_df['hashrate'] = raw_data_df['speed'] / raw_data_df['marketFactor'] * raw_data_df['miningFactor']
calc_df['hashrate unit'] = raw_data_df['displayMarketFactor'] + '/s'

calc_df['paying'] = raw_data_df['paying'] * raw_data_df['priceFactor'] / sats2btc
calc_df['paying unit'] = 'BTC/' + raw_data_df['displayPriceFactor'] + '/day'

calc_df['priceDownStep'] = raw_data_df['priceDownStep']
calc_df['priceDownStep unit'] = 'BTC/' + raw_data_df['displayPriceFactor'] + '/day'

price_down = calc_df['priceDownStep'] / raw_data_df['priceFactor'] / raw_data_df['paying'] * sats2btc
calc_df['priceDown'] = price_down
calc_df['priceDownPercent'] = round(abs(price_down) * 100, 2)

calc_df
````
